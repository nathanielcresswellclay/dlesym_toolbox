<h1>Debug GT Forecast</h1>

```python
import xarray as xr 
from matplotlib import pyplot as plt

def main(
    land_fcst: str,
    land_verif: str,
    plot_date: str,
    init_time: str,
    channel: str,
):

    print(f'Loading forecast from {land_fcst}')
    print(f'Loading verification from {land_verif}')
    land_fcst = xr.open_dataset(land_fcst)
    land_verif = xr.open_dataset(land_verif, engine='zarr')

    print(f'Fixing time coordinate')
    # fix the time coordinate
    land_fcst = land_fcst.assign_coords(dict(step=land_fcst.step.values + land_fcst.time.values[0]))

    print(f'Selecting verification and forecast')
    ndvi_test_verif = land_verif.targets.sel(time=plot_date, face=11, channel_out=channel)
    ndvi_test_fcst  = land_fcst[channel].sel(step=plot_date, face=11, time=land_fcst.time.values[0])

    # print values of the forecast and verification
    print(f'Forecast values: {ndvi_test_fcst.values}')
    print(f'Verification values: {ndvi_test_verif.values}')
    
    print(f'Plotting')
    fig, ax = plt.subplots(ncols=2, figsize=(10, 5))

    ndvi_test_verif.plot(ax=ax[0], cmap='viridis')
    ax[0].set_title('Verification')
    ndvi_test_fcst.plot(ax=ax[1], cmap='viridis')
    ax[1].set_title('Forecast')
    fig.suptitle(f'{channel} on {plot_date}')
    fig.tight_layout()
    
    print(f'Saved figure to debug_gt_forecast_{channel}_{plot_date}_{init_time}.png')
    fig.savefig(f'debug_gt_forecast_{channel}_{plot_date}_{init_time}.png')
```
